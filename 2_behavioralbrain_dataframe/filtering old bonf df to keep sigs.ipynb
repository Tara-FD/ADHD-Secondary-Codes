```python
import pandas as pd
```

```python
bonf = pd.read_csv('bonf.csv')
```

```python
bonf.head()
```

```python
sig_bonf = bonf[(bonf['rw'] <= 0.05) | (bonf['fa'] <= 0.05) | (bonf['rs'] <= 0.05) | (bonf['total'] <= 0.05)| (bonf['gb'] <= 0.05) | (bonf['rs'] <= 0.05)]
```

```python
sig_bonf
```

```python
df_all = pd.read_csv('ALL_DF_400.csv')
```

```python
df_all.head()
```

```python
lst_bonf = list(sig_bonf['Unnamed: 0'])
```

```python
df_bonferroni = pd.DataFrame()
```

```python
for column in df_all.columns:
    #print(columns)
    for i in range(len(lst_bonf)):
        if column == lst_bonf[i]:
            df_bonferroni[column] = df_all[column]
```

```python
df_bonferroni['Subject_ID'] = df_all['Subject_ID']
```

```python
df_bonferroni['sws_rs'] = df_all['sws_rs']
```

```python
df_bonferroni['sws_gb'] = df_all['sws_gb']
```

```python
df_bonferroni['sws_ms'] = df_all['sws_ms']
```

```python
df_bonferroni['sws_fa'] = df_all['sws_fa']
```

```python
df_bonferroni['sws_rw'] = df_all['sws_rw']
```

```python
df_bonferroni['sws_total'] = df_all['sws_total']
```

```python
df_bonferroni.head()
```

```python
first_column = df_bonferroni.pop('sws_total')
df_bonferroni.insert(0, 'sws_total', first_column)
```

```python
first_column = df_bonferroni.pop('sws_rw')
df_bonferroni.insert(0, 'sws_rw', first_column)
```

```python
first_column = df_bonferroni.pop('sws_fa')
df_bonferroni.insert(0, 'sws_fa', first_column)
```

```python
first_column = df_bonferroni.pop('sws_ms')
df_bonferroni.insert(0, 'sws_ms', first_column)
```

```python
first_column = df_bonferroni.pop('sws_gb')
df_bonferroni.insert(0, 'sws_gb', first_column)
```

```python
first_column = df_bonferroni.pop('sws_rs')
df_bonferroni.insert(0, 'sws_rs', first_column)
```

```python
first_column = df_bonferroni.pop('Subject_ID')
df_bonferroni.insert(0, 'Subject_ID', first_column)
```

```python
df_bonferroni.to_csv('2_df_bonferroni.csv')
```
